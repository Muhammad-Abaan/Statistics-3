##### **1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results**

```python
population = np.random.normal(100, 10, 10000)
sample = np.random.choice(population, 1000)

pop_mean = population.mean()
sam_mean = sample.mean()

zstat, pvalue = ztest(sample, value=pop_mean, alternative='larger')

if pvalue <= 0.05:
    print('Reject the null hypothesis')
else:
    print('Failed to reject the null hypothesis')
```

##### **2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python**

```python
population = np.random.normal(100, 10, 10000)
sample = np.random.choice(population, 1000)

pop_mean = population.mean()
pop_std = population.std()
sam_mean = sample.mean()
sam_len = len(sample)

zscore = (sam_mean-pop_mean)/(pop_std/np.sqrt(sam_len))
pvalue = 1-stats.norm.cdf(zscore)

if pvalue <= 0.05:
    print('Reject the null hypothesis')
else:
    print('Failed to reject the null hypothesis')
```

##### **4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot**

```python
population = np.random.normal(100, 10, 10000)
sample = np.random.choice(population, 1000)

pop_mean = population.mean()
pop_std = population.std()
sam_mean = sample.mean()
sam_len = len(sample)

zscore = (sam_mean - pop_mean) / (pop_std / np.sqrt(sam_len))
pvalue = 2 * (1 - stats.norm.cdf(abs(zscore)))

if pvalue <= 0.05:
    print('Reject the null hypothesis')
else:
    print('Failed to reject the null hypothesis')



x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')


plt.fill_between(x, y, where=(x <= -1.96) | (x >= 1.96), color='skyblue', alpha=0.5, label='Rejection Region (=0.05)') # zcritical->1.96


plt.axvline(zscore, color='red', linestyle='--', label=f'Z = {zscore:.2f}')


plt.axvline(-1.96, color='black', linestyle=':', linewidth=1)
plt.axvline(1.96, color='black', linestyle=':', linewidth=1)


plt.title("Two-Tailed Z-Test Visualization")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

##### **5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing**

```python
x = np.linspace(-4, 8, 1000)
null_dist = stats.norm(loc=0, scale=1)
alt_dist = stats.norm(loc=2, scale=1)

plt.plot(x, null_dist.pdf(x), label='Null Hypothesis', color='blue')
plt.plot(x, alt_dist.pdf(x), label='Alternative Hypothesis', color='green')

plt.fill_between(x, null_dist.pdf(x), where=(x > 1.645), color='red', alpha=0.4, label='Type I Error ()')
plt.fill_between(x, alt_dist.pdf(x), where=(x <= 1.645), color='purple', alpha=0.3, label='Type II Error ()')

plt.legend()
plt.title("Type I and Type II Errors")
plt.show()
```

##### **6. Write a Python program to perform an independent T-test and interpret the results**

```python
population = np.random.normal(100, 10, 10000)
sample1 = np.random.choice(population, 1000)
sample2 = np.random.choice(population, 1000)

tstat, pvalue = stats.ttest_ind(sample1, sample2, alternative='two_sided')

# Null Hypothesis (H):
# The mean of sample1 == mean of sample2

# Alternative Hypothesis (H):
# The mean of sample1  mean of sample2

if pvalue <= 0.05:
    print('Reject the null hypothesis')
else:
    print('Failed to reject the null hypothesis')
```

##### **9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance**

```python
sample = np.random.normal(80, 10, 50)
mean = np.mean(sample)
sem = stats.sem(sample)

# Upper CI = mean + (tcritical * SE) 
# Lower CI = mean - (tcritical * SE)

tcr = stats.t.ppf(1-0.025, len(sample)-1)

upper_ci = mean + tcr*sem
lower_ci = mean - tcr*sem

print(f"95% Confidence Interval: ({lower_ci:.2f}, {upper_ci:.2f})")
```

##### **19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA**

```python
# Check normality
for i, g in enumerate(groups):
    stat, p = stats.shapiro(g)
    print(f"Group {i+1} Normality p-value: {p:.4f}")

# Check equal variance
stat, p = stats.levene(*groups)  # *groups -> unpacks groups -> levene(group1, group2, group3)
print(f"Levene's test for equal variance p-value: {p:.4f}")
```

<u>**Shapiro-Wilk Test (Normality Test)**</u>

Used to check whether a sample comes from a **normal distribution**.

- **Null Hypothesis (H):** The data is **normally distributed**.
- **Alternative Hypothesis (H):** The data is **not normally distributed**.

Interpretation of p-value:

- If `p > 0.05`: **Fail to reject H**  Data **is normal**.
- If `p  0.05`: **Reject H**  Data **is not normal** (non-normal distribution).

---

<u>**Levenes Test (Equality of Variances)**</u>

Used to check if **multiple groups have equal variances**  important for t-tests and ANOVA.

- **Null Hypothesis (H):** All groups have **equal variance**.
- **Alternative Hypothesis (H):** At least one group has a **different variance**.

Interpretation of p-value:

- If `p > 0.05`: **Fail to reject H**  Variances are **equal**.
- If `p  0.05`: **Reject H**  Variances are **not equal**.

---

**Note:**
- For Independence
    - **Independence** means that each observation should be independent of the others.
    - This is a **design assumption**, not easily tested through statistics.

##### **20. Perform a two-way ANOVA (Factorial ANOVA) test using Python to study the interaction between two factors and visualize the results**

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

data = pd.DataFrame({
    'Score': [80, 85, 90, 75, 70, 95, 60, 65, 70, 85, 88, 90], # Score(Dependent Variable) - Numeric
    'Method': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'A', 'B', 'C'], # Method(Category): A,B,C -> 3 levels
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F']  # Gender(Category): M,F -> 2 levels
})

model = ols('Score ~ C(Method) + C(Gender) + C(Method):C(Gender)', data=data).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

sns.boxplot(x='Method', y='Score', hue='Gender', data=data)
plt.title('Two-way ANOVA Boxplot')
plt.show()
```

**Why Use `ols()` in Two-Way ANOVA?**

The `ols()` function (Ordinary Least Squares) is used to **fit a linear model** to the data.

In the context of ANOVA:

- ANOVA is fundamentally a **special case of linear regression**, where:
  - The dependent variable is **numeric** (e.g., `Score`)
  - The independent variables (factors) are **categorical** (e.g., `Method`, `Gender`)

- `ols()` allows us to write the model formula in a simple, R-style syntax:



- `C(...)` treats variables as **categorical factors**
- `+` adds each factor's main effect
- `:` includes the **interaction effect** between Method and Gender

- Once the model is fitted, we use `anova_lm()` to analyze the **variance partitioning**:
- It tells us how much of the total variability in the dependent variable is explained by each factor and their interaction.
- The output includes F-statistics and p-values to assess statistical significance.

---



- `C(Method)` tests the **main effect** of teaching method.
- `C(Gender)` tests the **main effect** of gender.
- `C(Method):C(Gender)` tests the **interaction** between method and gender.

---

**Interpretation of the ANOVA Table**

| Source                   | sum_sq     | df  | F        | PR(>F)   |
|--------------------------|------------|-----|----------|----------|
| C(Method)                | 418.17     | 2   | 1.62     | 0.274    |
| C(Gender)                | 114.08     | 1   | 0.88     | 0.383    |
| C(Method):C(Gender)      | 78.17      | 2   | 0.30     | 0.749    |
| Residual                 | 774.50     | 6   |         |         |

- **C(Method): p = 0.274**  No significant difference in scores between different methods.
- **C(Gender): p = 0.383**  No significant difference in scores between genders.
- **Interaction (Method  Gender): p = 0.749**  No evidence that the effect of method depends on gender.

In all cases, p-values are > 0.05, so **we fail to reject the null hypothesis** for each  meaning **none of the factors or their interaction has a significant effect** on Score in this small dataset.

---


<u>**One-Way ANOVA vs Two-Way ANOVA**</u>

**One-Way ANOVA**

- **Purpose**: Compare the means of multiple groups (based on one categorical factor).
- **Hypotheses**:
  - **H**: All group means are equal.
  - **H**: At least one group mean is different.

**Two-Way ANOVA**

- **Purpose**: Evaluate the effect of **two categorical independent variables (factors)** and their **interaction** on a continuous dependent variable.
- **Hypotheses**:
  1. **H (Method)**: Method has no effect on Score.
  2. **H (Gender)**: Gender has no effect on Score.
  3. **H (Interaction)**: There is no interaction between Method and Gender.

---

**Key Difference**

| Feature              | One-Way ANOVA              | Two-Way ANOVA                            |
|----------------------|----------------------------|------------------------------------------|
| Number of Factors    | One                        | Two or more                              |
| Interaction Analysis |  Not considered          |  Interaction between factors is tested |
| Structure            | Group mean comparison      | Regression-style formula                 |
| Output               | Group-level mean difference| Factor-level + interaction effects       |

---

##### **24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results**

```python
sample = [9, 11, 13, 12, 10, 9, 8, 10]
sample_size = len(sample)
sample_variance = np.var(sample, ddof=1)
hypothesized_variance = 4

chi_square_stat = (sample_size - 1) * sample_variance / hypothesized_variance
p_value = 1 - stats.chi2.cdf(chi_square_stat, df=sample_size - 1)

print("Chi-square Statistic:", chi_square_stat)
print("P-value:", p_value)
```

**Chi-Square Test for Variance (Using Numeric Data)**

This test checks whether the **variance** of a numeric sample matches a **hypothesized population variance**.

---

**Test Statistic Formula**

$$
\chi^2 = \frac{(n - 1) \cdot s^2}{\sigma_0^2}
$$


- $n$: Sample size  
- $s^2$: Sample variance  
- $\sigma_0^2$: Hypothesized population variance  


---

**Hypotheses (Right-Tailed)**

- **Null Hypothesis (H):** population variance  hypothesized variance  
- **Alternative Hypothesis (H):** population variance > hypothesized variance

For a **two-tailed** test, calculate both tails of the distribution.

---

**When to Use**

- When data is **continuous (numeric)**  
- To test if variability differs from a known or claimed value  
- Not to be confused with chi-square tests for **categorical data**

##### **25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups**

```python
success_a, nobs_a = 45, 100
success_b, nobs_b = 30, 100

count = np.array([success_a, success_b])
nobs = np.array([nobs_a, nobs_b])

stat, p_value = sm.stats.proportions_ztest(count, nobs)
print("Z-statistic:", stat)
print("P-value:", p_value)
```

- This is a two-proportion Z-test, used to compare whether the proportion of successes in two independent groups is significantly different.

    - Null Hypothesis (H): The two population proportions are equal
    - Alternative Hypothesis (H): The two proportions are not equal (by default, a two-tailed test)
---

Group A: 45% success rate

Group B: 30% success rate

You're testing if this 15% difference is statistically significant.

- Z-statistic: Measures how far apart the two proportions are, standardized

##### **26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results**

```python
data1 = np.random.normal(0, 1, 50)
data2 = np.random.normal(0, 2, 50)

var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)
f_stat = var1 / var2

df1, df2 = len(data1) - 1, len(data2) - 1
p_value = 1 - stats.f.cdf(f_stat, df1, df2)

print("F-statistic:", f_stat)
print("P-value:", p_value)

x = np.linspace(0.01, 5, 500)
plt.plot(x, stats.f.pdf(x, df1, df2), label='F-distribution')
plt.axvline(f_stat, color='red', linestyle='--', label='F-statistic')
plt.title('F-Test Visualization')
plt.legend()
plt.grid(True)
plt.show()
```
